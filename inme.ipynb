```python
import numpy as np
import pandas as pd
```

```python
veri= pd.read_excel('healthcare-dataset-stroke-data.xlsx')
```

```python
veri.head()
```

```python
silinecek_gender=[]
t=0
for i in range(veri.shape[0]):
    if veri.loc[i,'gender']== 'Other':
        t+=1
        silinecek_gender.append(i)
print(t)
silinecek_gender
```

```python
veri.drop(silinecek_gender,axis=0,inplace=True)
```

```python
veri.reset_index(inplace=True)
```

```python
kisi= veri.shape[0]
oz= veri.shape[1]
```

```python
gender_num= np.zeros((kisi,1)).astype(int)
for i in range(kisi):
    if veri.loc[i,'gender']=='Female':
        gender_num[i]= 1
gender_num
```

```python
gender_num= pd.DataFrame(gender_num,columns=['gender_num'])
veri= pd.concat([gender_num, veri], axis=1)
```

```python
veri.drop('gender', axis=1, inplace=True)
```

```python
veri.drop('index', axis=1, inplace=True)
```

```python
veri.head()
```

```python
married_num=np.zeros(kisi).astype(int)
for i in range(kisi):
    if veri.loc[i,'ever_married']=='Yes':
        married_num[i]= 1
married_num= pd.DataFrame(married_num, columns=['married_num'])
veri= pd.concat([married_num, veri], axis=1)
veri.head()
```

```python
veri.drop('ever_married', axis=1, inplace=True)
```

```python
veri.head()
```

```python
veri['work_type'].value_counts()
work_dict= {'Private':0, 'Self-employed':1, 'children':2, 'Govt_job':3, 'Never_worked':4}
work_num= np.zeros((kisi,5)).astype(int)

for i in range(kisi):
    anlik_num= work_dict[veri.loc[i, 'work_type']]
    work_num[i,anlik_num]= 1
work_num

work_num= pd.DataFrame(work_num, columns=['work_1', 'work_2', 'work_3', 'work_4', 'work_5'])
veri= pd.concat([work_num,veri], axis=1)
veri.head()
```

```python
veri.drop('work_type', axis=1, inplace=True)
```

```python
veri.head()
```

```python
veri.value_counts('Residence_type')
```

```python
residence_num= np.zeros(kisi).astype(int)
for i in range(kisi):
    if veri.loc[i, 'Residence_type']== 'Urban':
        residence_num[i]=1
residence_num
residence_num=pd.DataFrame(residence_num, columns=['residence_num'])
veri=pd.concat([residence_num, veri], axis=1)
veri.head()
```

```python
veri.drop('Residence_type', axis=1, inplace=True)
```
